## Preprocessing

```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("https://static.bc-edx.com/data/dla-1-2/m21/lms/starter/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
application_df2 = application_df.drop(['EIN','NAME'],1)
application_df2.head()
```

```python
# Determine the number of unique values in each column.
application_df2.nunique()
```

```python
# Look at APPLICATION_TYPE value counts for binning
application_df2.APPLICATION_TYPE.value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
value_counts = application_df2.APPLICATION_TYPE.value_counts().rename_axis('APPLICATION_TYPE').reset_index(name='counts')
value_counts2 = value_counts[value_counts['counts'] < 200]
application_types_to_replace = value_counts2['APPLICATION_TYPE']

# Replace in dataframe
for app in application_types_to_replace:
    application_df2['APPLICATION_TYPE'] = application_df2['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
application_df2['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
application_df2.CLASSIFICATION.value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1
vc = application_df2.CLASSIFICATION.value_counts()
print(vc[vc > 1])
```

```python
# Choose a cutoff value and create a list of classifications to be replaced
# use the variable name `classifications_to_replace`
value_counts = application_df2.CLASSIFICATION.value_counts().rename_axis('CLASSIFICATION').reset_index(name='counts')
value_counts2 = value_counts[value_counts['counts'] < 1200]
classifications_to_replace = value_counts2['CLASSIFICATION']

# Replace in dataframe
for cls in classifications_to_replace:
    application_df2['CLASSIFICATION'] = application_df2['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
application_df2['CLASSIFICATION'].value_counts()
```

```python
application_df2.columns
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
features = pd.get_dummies(application_df2[['APPLICATION_TYPE', 'AFFILIATION', 'CLASSIFICATION', 'USE_CASE',
       'ORGANIZATION', 'INCOME_AMT', 'SPECIAL_CONSIDERATIONS']])
features['STATUS'] = application_df2['STATUS']
features['ASK_AMT'] = application_df2['ASK_AMT']
features.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = application_df2['IS_SUCCESSFUL']

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(features, y)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn = tf.keras.models.Sequential()
nn.add(tf.keras.layers.Dense(units=200, activation="relu", input_dim=43))

# First hidden layer
nn.add(tf.keras.layers.Dense(units=200, activation="relu"))

nn.add(tf.keras.layers.Dropout(0.1))

# Second hidden layer
nn.add(tf.keras.layers.Dense(units=200, activation="tanh"))

nn.add(tf.keras.layers.Dropout(0.1))

nn.add(tf.keras.layers.Dense(units=200, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn.summary()
```

```python
# Compile the model
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
train = nn.fit(X_train_scaled, y_train, epochs=40)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```

```python
# Export our model to HDF5 file
nn.save('my_model.h5')
```
